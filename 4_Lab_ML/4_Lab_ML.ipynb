# Регрессия. Определение
Термин «регрессия» был придуман сэром Фрэнсисом Гальтоном в конце 19 века. Он использовал этот термин для описания феномена детского роста, имеющего тенденцию к регрессии к среднему значению населения, что означает, что родители выше среднего, как правило, имеют детей, которые ближе к среднему росту, а родители ниже среднего, как правило, имеют детей, которые ближе к среднему росту.

Что такое регрессионный анализ?
Регрессионный анализ — это статистический процесс оценки отношений между зависимыми переменными или переменными-критериями и одной или несколькими независимыми переменными или предикторами. Регрессионный анализ обычно используется, когда мы имеем дело с набором данных, который имеет целевую переменную в виде непрерывных данных. Регрессионный анализ объясняет изменения в критериях или изменения в выбранных предикторов. Условное ожидание критериев основано на предикторах, где среднее значение зависимых переменных задано при изменении независимых переменных. Регрессионный анализ используется в трех основных направлениях использования: определение силы предикторов, прогнозирование эффекта и прогнозирование трендов.

Типы методов регрессии
Существует несколько типов методов регрессии, каждый из которых подходит для различных типов данных и различных типов отношений.
В регрессии используются два типа переменных: независимые переменные и зависимая переменная. Независимые переменные являются входными данными для регрессионной модели, в то время как зависимая переменная — это выходные данные, которые модель пытается предсказать.

Основными методами регрессии являются:

1. Линейная регрессия
1. Полиномиальная регрессия
1. Ступенчатая регрессия
1. Регрессия дерева принятия решений
1. Регрессия случайного леса
1. Регрессия опорных векторов
1. Регрессия хребта
1. Регрессия лассо
1. Регрессия ElasticNet
1. Байесовская линейная регрессия



Регрессия является мощным инструментом для понимания и прогнозирования взаимосвязей между переменными. Он используется в самых разных областях, включая финансы, экономику, маркетинг и медицину.

Существует два основных типа регрессии —  линейная регрессия и логистическая регрессия. Линейная регрессия используется для прогнозирования непрерывного числового результата. Логистическая регрессия используется для прогнозирования двоичного категориального результата, то есть логистическая регрессия - это классификатор!


# Метрики регрессии

Все эти метрики являются по сути функциями потерь (**loss functions**), потому что мы пытаемся их минимизировать. Кроме, конечно, метрики R2, которую мы пытаемся максимизировать



## 2.1 Средняя абсолютная ошибка - Mean Absolute Error (MAE)

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

Средняя абсолютная ошибка представляет собой среднее значение абсолютной разницы между фактическими и прогнозируемыми значениями в наборе данных. Она измеряет среднее значение остатков.




## 2.2 Среднеквадратическая ошибка - Mean Squared Error (MSE)

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

Среднеквадратическая ошибка представляет собой среднее квадратичной разницы между исходными и прогнозируемыми значениями в наборе данных. Она измеряет дисперсию остатков.




## 2.3 Квадратный корень из среднеквадратической ошибки - Root Mean Squared Error (RMSE)

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

RMSE измеряет стандартное отклонение остатков.

## 2.4 R-squared (Коэффициент детерминации)

1. **SST** (или TSS)

Сумма квадратов всего/общая сумма квадратов (SST или TSS) — это квадрат разности между наблюдаемой зависимой переменной и ее средним значением.

2. **SSR** (или RSS)

Регрессия суммы квадратов (SSR или RSS) — это сумма разностей между прогнозируемым значением и средним значением зависимой переменной.

3. **SSE** (или ESS)

Ошибка суммы квадратов (SSE или ESS) — это разница между наблюдаемым значением и прогнозируемым значением.


Коэффициент детерминации или R-квадрат представляет собой долю дисперсии в зависимой переменной, которая объясняется линейной регрессионной моделью. Когда R² высок, это означает, что регрессия может охватить большую часть вариации в наблюдаемых зависимых переменных. Вот почему мы можем сказать, что регрессионная модель работает хорошо, когда R² высок.

$$R^2 = 1- \frac {SSR}{SST}$$

Это безмасштабная оценка, т. е. независимо от того, являются ли значения малыми или большими, значение R-квадрат будет меньше единицы. Одно из заблуждений относительно регрессионного анализа заключается в том, что низкое значение R-квадрат всегда плохо. Например, некоторые наборы данных или области изучения имеют изначально большее количество необъяснимой вариации. В этом случае значения R-квадрат, естественно, будут ниже. ## 2.5 Скорректированный R квадрат

$$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$$

**Скорректированный R квадрат** — это модифицированная версия R2, скорректированная с учетом количества независимых переменных в модели, и она всегда будет меньше или равна R². В приведенной выше формуле n — это количество наблюдений в датасете, а k — это количество независимых переменных.

## 2.6 Перекрестная проверка R2

Перекрестная проверка — это процедура повторной выборки, используемая для оценки моделей машинного обучения на ограниченной выборке данных. Это популярный метод, поскольку он прост для понимания и обычно приводит к менее предвзятой или менее оптимистичной оценке мастерства модели, чем другие методы, такие как простое разделение на обучение/тест.

Процедура имеет один параметр, называемый k, который относится к количеству групп, на которые следует разделить данную выборку данных. Таким образом, процедура часто называется k-кратной перекрестной проверкой. Когда выбрано конкретное значение для k, его можно использовать вместо k в ссылке на модель, например, k=10 становится 10-кратной перекрестной проверкой.

**Перекрестная проверка R2** — это медианное значение или R2, взятое из процедуры перекрестной проверки.

## 2.7 Метрики оценки регрессии — заключение

Как RMSE, так и R-квадрат количественно определяют, насколько хорошо модель линейной регрессии соответствует набору данных. При оценке того, насколько хорошо модель соответствует набору данных, полезно вычислить как RMSE, так и значение R2, поскольку каждая метрика говорит нам что-то свое.

* **RMSE** сообщает нам типичное расстояние между прогнозируемым значением, полученным с помощью модели регрессии, и фактическим значением.

* **R2** сообщает нам, насколько хорошо переменные-предикторы могут объяснить вариацию в целевой переменной.

Добавление большего количества независимых переменных или предикторов в модель регрессии, как правило, увеличивает значение R2, что побуждает создателей модели добавлять еще больше переменных. **Скорректированный R2** используется для определения того, насколько надежна корреляция и насколько она определяется добавлением независимых переменных. Он всегда ниже, чем R2.

```python
# сразу напишем код функции, которая будет вычислять все нужные метрики

def Reg_Models_Evaluation_Metrics (model,X_train,y_train,X_test,y_test,y_pred):
    cv_score = cross_val_score(estimator = model, X = X_train, y = y_train, cv = 10)

    # Вычисление "скорректировнного R2"
    # Сначала вычислим коэффициент детерминации R2
    r2 = model.score(X_test, y_test)
    # Размер тестовой выборки
    n = X_test.shape[0]
    # Количество признаков
    p = X_test.shape[1]
    # Вычислим "Скорректировнный R2"
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
    R2 = model.score(X_test, y_test)
    CV_R2 = cv_score.mean()

    print(f'RMSE:  {round(RMSE,4)}\n  R2: {round(R2,4)}\n Adjusted R2: {round(adjusted_r2, 4)}\n Cross Validated R2: {round(cv_score.mean(),4)}')
    return R2, adjusted_r2, CV_R2, RMSE
```

# Линейная регрессия

Линейная регрессия используется для предиктивного анализа. Линейная регрессия — это линейный подход к моделированию взаимосвязи между целевой переменной и множественными предикторами (независимыми переменными). Линейная регрессия фокусируется на условном распределении вероятностей ответа с учетом значений предикторов. При линейной регрессии существует опасность переобучения. Формула линейной регрессии выглядит следующим образом:

y = θx + b

где

θ – веса (параметры) модели

b –  смещение.

Это самая простая форма регрессионного анализа, которая используется для моделирования линейной связи между одной зависимой переменной и одной или несколькими независимыми переменными.

# Полиномиальная регрессия

Это расширение линейной регрессии и используется для моделирования нелинейной связи между зависимой переменной и независимыми переменными. Здесь синтаксис также остается прежним, но теперь во входные переменные мы включаем некоторые полиномиальные или более высокие члены некоторых уже существующих признаков. Линейная регрессия смогла подогнать только линейную модель к имеющимся данным, но с помощью полиномиальных признаков мы можем легко подогнать некоторую нелинейную связь между целевыми и входными признаками.

Полиномиальная регрессия — это форма линейной регрессии, в которой связь между независимой переменной x и зависимой переменной y моделируется как полином n-й степени . Полиномиальная регрессия соответствует нелинейной связи между значением x и соответствующим условным средним значением y, обозначаемым E(y | x).

Почему полиномиальная регрессия?
Полиномиальная регрессия — это тип регрессионного анализа, используемый в статистике и машинном обучении, когда связь между независимой переменной (вход) и зависимой переменной (выход) не является линейной. В то время как простая линейная регрессия моделирует связь как прямую линию, полиномиальная регрессия обеспечивает большую гибкость, подгоняя полиномиальное уравнение к данным.

Когда связь между переменными лучше представлена ​​кривой, а не прямой линией, полиномиальная регрессия может улавливать нелинейные закономерности в данных.

Как работает полиномиальная регрессия?
Если мы внимательно посмотрим, то поймем, что для перехода от линейной регрессии к полиномиальной регрессии нам просто нужно добавить члены более высокого порядка зависимых признаков в пространстве признаков. Иногда это также называют проектированием признаков , но не совсем так.

Если связь нелинейная, модель полиномиальной регрессии вводит полиномиальные члены более высокой степени.

Общая форма уравнения полиномиальной регрессии степени n имеет вид:

$y=β_0+β_1x+β_2x^2+…+β_nx^n +ϵ$     

где,

y — зависимая переменная.
x — независимая переменная.
$β_0,β_1,…,β_n$    — коэффициенты полиномиальных членов.
n — степень многочлена.
$ϵ$ представляет собой ошибочный член.
Основная цель регрессионного анализа — смоделировать ожидаемое значение зависимой переменной y в терминах значения независимой переменной x. В простой линейной регрессии мы использовали следующее уравнение —

$у = а + Ьх + е$
Здесь y — зависимая переменная, a — y-отрезок, b — наклон, а e — коэффициент ошибки.

Во многих случаях эта линейная модель не сработает. Например, если мы проанализируем производство химического синтеза с точки зрения температуры, при которой происходит синтез, в таких случаях мы используем квадратичную модель.
$у = а + b_1x + b_2x^2 + е$    

Здесь,

y — зависимая переменная от x
a — это точка пересечения с осью Y, а e — это частота ошибок.

В общем случае мы можем смоделировать его для n-го значения.

$у = а + b_1x + b_2x^2 +....+ b_nx^n$    

Поскольку функция регрессии линейна относительно неизвестных переменных, то эти модели линейны с точки зрения оценки. Следовательно, с помощью метода наименьших квадратов можно вычислить значение отклика (y).

Включая члены более высокой степени (квадратичные, кубические и т. д.), модель может улавливать нелинейные закономерности в данных.

Выбор степени полинома ( n ) является важнейшим аспектом полиномиальной регрессии. Более высокая степень позволяет модели лучше соответствовать обучающим данным, но это также может привести к переобучению, особенно если степень слишком высока. Поэтому степень следует выбирать на основе сложности базовой связи в данных.
Модель полиномиальной регрессии обучается находить коэффициенты , которые минимизируют разницу между прогнозируемыми значениями и фактическими значениями в обучающих данных.
После обучения модели ее можно использовать для прогнозирования новых, невиданных данных. Полиномиальное уравнение фиксирует нелинейные закономерности, наблюдаемые в обучающих данных, что позволяет модели обобщаться до нелинейных отношений.

При работе с полиномиальной регрессией мы сталкиваемся с проблемой переобучения. Это происходит из-за того, что при увеличении порядка полиномиальной регрессии для достижения все более высокой производительности модель переобучается на данных и не работает на новых точках данных.

# Регрессия случайного леса
Random Forest — это метод ансамбля, способный выполнять задачи как регрессии, так и классификации с использованием нескольких деревьев решений и метода, называемого Bootstrap and Aggregation, широко известного как bagging. Основная идея заключается в том, чтобы объединить несколько деревьев решений для определения конечного результата, а не полагаться на отдельные деревья решений.

Random Forest имеет несколько деревьев решений в качестве базовых моделей обучения. Мы случайным образом выполняем выборку строк и признаков из набора данных, формируя образцы наборов данных для каждой модели. Эта часть называется Bootstrap.

Регрессия случайного леса — это универсальный метод машинного обучения для прогнозирования числовых значений. Он объединяет прогнозы нескольких деревьев принятия решений для уменьшения переобучения и повышения точности. Библиотеки машинного обучения Python упрощают реализацию и оптимизацию этого подхода.

Ансамблевое обучение
Ансамблевое обучение — это метод машинного обучения, который объединяет прогнозы из нескольких моделей для создания более точного и стабильного прогноза. Это подход, который использует коллективный интеллект нескольких моделей для повышения общей производительности системы обучения.

Виды ансамблевых методов
Существуют различные типы методов ансамблевого обучения, в том числе:

Бэггинг (Bootstrap Aggregating): Этот метод включает в себя обучение нескольких моделей на случайных подмножествах обучающих данных. Затем прогнозы отдельных моделей объединяются, как правило, путем усреднения.
Бустинг (Boosting): Этот метод включает в себя обучение последовательности моделей, где каждая последующая модель фокусируется на ошибках, допущенных предыдущей моделью. Прогнозы объединяются с помощью схемы взвешенного голосования.
Накопление (Stacking): Этот метод предполагает использование прогнозов из одного набора моделей в качестве входных признаков для другой модели. Окончательный прогноз делается моделью второго уровня.

Случайный лес — это метод ансамблевого обучения, который объединяет прогнозы из нескольких деревьев решений для получения более точного и стабильного прогноза. Это тип алгоритма контролируемого обучения, который можно использовать как для задач классификации, так и для задач регрессии.

Каждое дерево решений имеет высокую дисперсию, но когда мы объединяем их все параллельно, результирующая дисперсия низкая, поскольку каждое дерево решений идеально обучается на этих конкретных выборочных данных, и, следовательно, результат зависит не от одного дерева решений, а от нескольких деревьев решений. В случае задачи классификации окончательный результат получается с помощью классификатора с большинством голосов. В случае регрессионной задачи конечный результат является средним значением всех выходов. Эта часть называется агрегированием.

Регрессия случайного леса в машинном обучении — это ансамблевый метод, способный выполнять задачи как регрессии, так и классификации с использованием нескольких деревьев решений и метода под названием Bootstrap and Aggregation, широко известного как бэггинг. # Регрессия ElasticNet

Линейная регрессия страдает от переобучения и не может работать с коллинеарными данными. Когда в наборе данных много признаков и даже некоторые из них не имеют отношения к прогностической модели. Это делает модель более сложной из-за слишком неточного прогнозирования на тестовом наборе (или переобучения). Такая модель с высокой дисперсией не обобщает на новых данных. Итак, чтобы решить эти проблемы, мы включили нормировку как L-2, так и L-1, чтобы получить преимущества как Ridge, так и Lasso одновременно. Результирующая модель обладает большей предсказательной силой, чем Лассо. Он выполняет выбор признаков, а также упрощает гипотезу.

Линейная регрессия - это метод второго порядка с моделью регуляризации Elastic Net от штрафа L1 Лассо и штрафа L2 от Ridge Methods. Первое наказание, L1 или Лассо, делает некоторые коэффициенты равными нулю, потому что алгоритм не позволяет использовать это значение, в то время как второе, L2 или Ridge, уменьшает коэффициенты до нуля, не заставляет их быть равными нулю.

Это совокупность этих штрафов с параметром регуляризации, обозначенным как альфа, который отображает все ограниченные силы, а коэффициент смешивания, известный как l1_ratio, определяет степень штрафов L1 и L2. Целевая функция Elastic Net может быть записана следующим образом: Целевая функция Elastic Net может быть записана как:

$minimize: 1/(2 * n_{samples}) * ||y – Xw||^2_2 + alpha * l1_{ratio} * ||w||_1 + 0.5 * alpha * (1 – l1_{ratio}) * ||w||^2_2$

где
$y$ — целевая переменная,

$X$ — входные данные,

$w$ — вектор коэффициентов,

$n_{samples}$ — количество выборок,

$alpha$ — сила регуляризации,

$l1_{ratio}$ — параметр регуляризации.


Регуляризация Elastic Net сочетает в себе сильные стороны методов регуляризации Lasso и Ridge:

Как и Лассо, он может обрабатывать многомерные данные и выполнять выбор признаков, сводя некоторые коэффициенты точно к нулю.

Как и Ridge, он может обрабатывать мультиколлинеарность (высококоррелированные объекты) и уменьшать коэффициенты до нуля.


###Ключевые особенности ElasticNet

**Комбинация штрафов L1 и L2**

 ElasticNet линейно комбинирует штрафы L1 и L2, что помогает в изучении разреженной модели с небольшим количеством ненулевых весов, сохраняя при этом свойства регуляризации ридж-регрессии.

**Обработка мультиколлинеарности**

 ElasticNet особенно полезен при работе с наборами данных, которые имеют несколько коррелированных объектов. В отличие от Лассо, которое может случайным образом выбрать один объект из группы коррелированных признаков, ElasticNet имеет тенденцию выбирать все коррелированные объекты вместе.

**Гиперпараметры**

alpha: Контролирует общую силу регуляризации.

l1_ratio: Определяет сочетание штрафов L1 и L2. Значение 0 соответствует регрессии Риджа, 1 — Лассо, а значения между ними — сочетание того и другого.



```python
from sklearn.linear_model import ElasticNet

# Create an elastic net regression model
modelEN = ElasticNet(alpha=0.05, l1_ratio=0.0)

# Fit the model to the data
modelEN.fit(X_train, y_train)

# Predict the response for a new data point
y_en_pred = modelEN.predict(X_test)

_ = Reg_Models_Evaluation_Metrics(modelEN, X_train, y_train, X_test, y_test, y_en_pred)
```
