```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # устанавливаем seaborn по умолчанию для отрисовки графиков
from sklearn.preprocessing import LabelEncoder # кодировка символов в числовые значения
from sklearn.preprocessing import MinMaxScaler # нормализация
```

```python
# загрузка данных
train = pd.read_csv('/content/mushrooms.csv')
```

```python
# вывод всех столбцов

train_cols_ru = ["класс", "форма шляпки", "поверхность шляпки", "цвет шляпки", "синяки", "запах",
"прикрепление пластинок", "расстояние между пластинками", "размер пластинок", "цвет пластинок",
"форма стебля", "стебель-корень", "поверхность стебля над кольцом",
"поверхность стебля под кольцом", "цвет стебля над кольцом",
"цвет стебля под кольцом", "тип оболочки", "цвет оболочки", "количество колец",
"тип кольца", "цвет спор", "популяция", "среда обитания"]


translation_dict = dict(zip(train.columns, train_cols_ru))

for key, value in translation_dict.items():
    print(f"{key}: {value}")
```

```python
# все категории таблицы
for column in train.columns:
    unique_values = train[column].unique()
    print(f"Уникальные значения в столбце '{column}': {unique_values}")
```

```python
train.head()
```

```python
# распределение всех грибов по форме шляпы
# колоколообразная=b, коническая=c, выпуклая=x, плоская=f, выпуклая с бугорком=k, вогнутая=s
# больше всех грибов с выпуклой формой (значение - x)
# меньше всех - с конической формой (значение - c)
sns.countplot(x=train["cap-shape"]);
```

```python
# распределение всех грибов по форме шляпы
# волокнистая=f, бороздчатая=g, чешуйчатая=y, гладкая=s
sns.countplot(x=train["cap-surface"]);
```

```python
# распределение всех грибов по цвету шляпки
# коричневый=n, желтовато-коричневый=b, коричный=c, серый=g, зелёный=r, розовый=p, фиолетовый=u, красный=e, белый=w, жёлтый=y
sns.countplot(x=train["cap-color"]);
```

```python
# распределение ядовитых грибов внутри каждой группы, которые отличаются формой шляпы
# колоколообразная=b, коническая=c, выпуклая=x, плоская=f, выпуклая с бугорком=k, вогнутая=s
# для выпуклой и плоской форм шляп доли ядовитых и съедобных грибов примерно равны (значения x и f)
# для колоколообразной и выпуклой с бугорком форм шляп доли ядовитых и съедобных сильно разнятся (значения b и k)
sns.countplot(data=train, x="cap-shape", hue='class');
```

```python
# распределение ядовитых грибов внутри каждой группы, которые отличаются поверхностью шляпы
# волокнистая=f, бороздчатая=g, чешуйчатая=y, гладкая=s
sns.countplot(data=train, x="cap-surface", hue='class');
```

```python
# распределение ядовитых грибов внутри каждой группы, которые отличаются цветом шляпки
# коричневый=n, желтовато-коричневый=b, коричный=c, серый=g, зелёный=r, розовый=p, фиолетовый=u, красный=e, белый=w, жёлтый=y
sns.countplot(data=train, x="cap-color", hue='class');
```

```python
# распределение ядовитых грибов внутри каждой группы, которые отличаются запахом
# миндальный=f, анисовый=l, креозотовый=c, рыбный=y, неприятный=f, затхлый=m, отсутствует=n, резкий=p, пряный=s
sns.countplot(data=train, x="odor", hue='class');
```

```python
# преобразование категориальных меток в числовые значения
labelencoder = LabelEncoder()

for col in train.columns:
    train[col] = labelencoder.fit_transform(train[col])
```

```python
# все категории таблицы (числовые)
for column in train.columns:
    unique_values = train[column].unique()
    print(f"Уникальные значения в столбце '{column}': {unique_values}")
```

```python
# нормализация мин-макс
scaler_mm = MinMaxScaler()

train[['mm-cap-shape','mm-cap-surface', 'mm-cap-color', 'mm-odor', 'mm-gill-color', 'mm-stalk-root', 'mm-stalk-surface-above-ring', 'mm-stalk-surface-below-ring', 'mm-stalk-color-above-ring', 'mm-stalk-color-below-ring', 'mm-veil-color', 'mm-ring-number', 'mm-ring-type', 'mm-spore-print-color', 'mm-population', 'mm-habitat']] = scaler_mm.fit_transform(train[['cap-shape','cap-surface', 'cap-color', 'odor', 'gill-color', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']])
train.head()
```

```python
train_norm = train.copy()

# удаление старых столбцов после нормализации
# также удаляется veil-type (тип оболочки), т.к. данный критерий линейно зависим от остальных параметров, это было заметно по тепловой карте
columns_to_drop = ['veil-type', 'cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-color', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']
train_norm.drop(columns=columns_to_drop, inplace=True)

train_norm.head()
```
